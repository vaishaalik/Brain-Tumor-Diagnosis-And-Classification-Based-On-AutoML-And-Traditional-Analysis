```python
from skimage.color import rgb2gray
from skimage.io import imread, imsave
from skimage.filters import threshold_multiotsu
from skimage import img_as_ubyte
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image

origin_path = 'D:/brainTumorClassification/Testing/glioma_tumor/'
dest_path = "./gray_scale/testing/glioma_tumor/"
```

```python
for filename in os.listdir(origin_path):
    img = imread(origin_path + filename)
    image_gray = rgb2gray(img)
    thresholds = threshold_multiotsu(image_gray, classes = 5)
    regions = np.digitize(image_gray, bins=thresholds)
    output = img_as_ubyte(regions)
    plt.imsave(dest_path + filename, output)
```
