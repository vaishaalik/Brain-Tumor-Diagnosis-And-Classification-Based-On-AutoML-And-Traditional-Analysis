```python
def otsuInv(path):
    i=0
    for jpgfile in os.listdir(path):
        i=i+1
        img = cv2.imread(path+jpgfile)
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        ret2,Otsu_inv = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
        dst = '/content/drive/MyDrive/BrainTumor/InverseOtsu/Training/glioma_tumor/'+jpgfile # alter this as required
        ret = cv2.imwrite(dst,Otsu_inv)
        print("Augmenting...",ret)
```
